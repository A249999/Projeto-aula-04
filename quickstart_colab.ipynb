```python
from transformers import pipeline

modelo = pipeline("text-generation", model="EleutherAI/gpt-neo-125M")

def gerar_resposta_criativa(prompt):
  resposta = modelo(prompt, max_length=50, num_return_sequences=1, no_repeat_ngram_size=2)[0]['generated_text']
  return resposta

# Loop de interação
while True:
  prompt = input("Você: ")
  resposta = gerar_resposta_criativa(prompt)
  print("Chatbot:", resposta)
```
